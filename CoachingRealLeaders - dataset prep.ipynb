```python
import requests
from bs4 import BeautifulSoup
from pathlib import Path

for s_num, season_episodes in episodes.items():
    for e_num, url in season_episodes.items():
        episode_file = Path('transcripts', f'{s_num}_{e_num}.txt')
        response = requests.get(url)
        html_content = response.text

        try: 
            soup = BeautifulSoup(html_content, 'html.parser')
            transcript_div = soup.find('div', {'class': 'podcast-tabs__content'})
            paragraphs = transcript_div.findAll('p')
            with open(episode_file, 'w') as f:
                for paragraph in paragraphs:
                    f.write(paragraph.text + '\n\n')
            
            print(f"Transcript successfully saved to {episode_file}")
        except: 
            print(f"Could not load/find/parse {episode_file}")
```

# Process episode and create an Instruct dataset

```python
from os import listdir
from os.path import isfile
from pathlib import Path
import numpy as np
import json

data_in = 'transcripts'
json_dir = 'transcripts_json'

HEADER_CUE = "Today’s guest"
COACH = "Muriel Wilkins"
COACH_CUE = COACH.upper() + ":"
CONTEXT_HEADER = "You are professional executive coach working with clients that are hitting a bump in the road, you guide them with questions into provoking thoughts and reactions. For this session we have a new guest. "
REPLACE_MAP = {"Coaching Real Leaders": "a coaching session", "the HBR Presents Network": "the executive coaching sessions"}
COACH_GUIDE = "As a coach, you are only allowed to iterate using open-ended questions, better one question at a time and using more questions as the conversation flows"
DEBUG = False

    
def remove_non_dialogue_lines(lines):
    lines = [l for l in lines if len(l) > 0]
    lines = [l for l in lines if l[-1] != ":"]  # e.g. "Further reading:"
    return lines

def find_guest_name(lines):
    """ Implement a majority vote """
    guest_names = []
    for l in lines:
        if l != "":
            if l.find(":") != -1:
                if l[:len(COACH_CUE)] != COACH_CUE:
                    guest_names.append(l[:l.find(":")])
    names, freq = np.unique(guest_names, return_counts=True)
    return names[np.argmax(freq)]

def get_context(lines):
    if lines[0].find(":") == -1:
        return lines[0]

def remove_header(s):
    header_removed = False
    start_pos = s.find(HEADER_CUE)
    if start_pos != -1:
        s = s[:s.find(":")+1] + " " + s[start_pos:]
        header_removed = True
    return s, header_removed

def remove_promo(c_lines):
    return c_lines[:-1]  # last line is social promo

def sanitize(s, guest):
    s = s.replace(COACH, "coach")
    s = s.replace(COACH.split(' ')[0], "coach")
    s = s.replace(guest.capitalize(), "guest")
    for k, v in REPLACE_MAP.items():
        s = s.replace(k, v)
    s, _ = remove_header(s)
    if s[0] == " ":
        s = s[1:]
    return s

def process_episode(lines):
    conv_lines = []
    guest_name = find_guest_name(lines)
    guest_cue = guest_name.upper() + ":"
    context = get_context(lines)
    conv_lines.append({"role": "system", "content": CONTEXT_HEADER + sanitize(context, guest_name)})
    conv_lines.append({"role": "system", "content": COACH_GUIDE})
    for l in lines:
        l = l.replace(" :", ":")
        if COACH_CUE in l:
            conv_lines.append({"role": "assistant", "content": sanitize(l.replace(COACH_CUE, ""), guest_name)})
        if guest_cue in l:
            conv_lines.append({"role": "user", "content": sanitize(l.replace(guest_cue, ""), guest_name)})
        # debug line
        if DEBUG and (COACH_CUE not in l) and (guest_cue not in l):
            print(l)
    return remove_promo(conv_lines)


total_instructs = 0
for filename in listdir(data_in):
    fn = Path(data_in, filename)
    if isfile(fn):
        with open(fn) as f:
            try:
                mem_file = f.read()
                lines = mem_file.split("\n\n")
                lines = remove_non_dialogue_lines(lines)
                conv_lines = {"messages": process_episode(lines)}

                with open(Path(json_dir, filename.replace("txt", "json")), "w", encoding="utf-8") as jf:
                    json.dump(conv_lines, jf, ensure_ascii=False)
                
                print(f"Processed {fn} length = {len(conv_lines['messages'])}")
                total_instructs += len(conv_lines["messages"])

            except UnicodeDecodeError:  # e.g. DS_Store file
                print(f"Skipping file {fn}")
            
total_instructs
```
